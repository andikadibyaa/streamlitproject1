### Cleaning Data

#### Penanganan Outlier pada Kedua Dataset

```python
# Drop Outliers
def remove_outliers(df):
    # Calculate the IQR for each column
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Drop the rows containing outliers for all columns
    df = df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]
    return df

print(f"data_day shape before removing outliers: {data_day.shape}")
data_day = remove_outliers(data_day)
print(f"data_day shape after removing outliers: {data_day.shape}")

print(f"data_hour shape before removing outliers: {data_hour.shape}")
data_hour = remove_outliers(data_hour)
print(f"data_hour shape after removing outliers: {data_hour.shape}")
```

Dari output di atas, dapat dilihat bahwa terdapat penghilangan beberapa rows dikarenakan terdapat outlier pada rows tersebut. Hal ini juga menandakan bahwa outlier sudah teratasi.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Kecenderungan Casual dan Registered Renter terhadap Pemilihan Jam

#### Melihat Perbandingannya dengan Line Chart Plot

```python
# Mendefinsikan Data
hr_labels = [f'Hour {i}' for i in range(24)]
data_hour['hr_name'] = data_hour['hr'].map(dict(enumerate(hr_labels)))
data_hour_grouped = data_hour.groupby('hr_name')
casual_by_hr = data_hour_grouped['casual'].sum()
registered_by_hr = data_hour_grouped['registered'].sum()
#plot line chart
sns.set(style="whitegrid")
plt.figure(figsize=(20, 8)) 
plt.rcParams['axes.facecolor'] = "#F0F0F0"
casual_color = '#2C599D'
registered_color = '#FF5733'
plt.plot(casual_by_hr.index, casual_by_hr, label='Casual', color=casual_color, marker='o', linewidth=2)
plt.plot(registered_by_hr.index, registered_by_hr, label='Registered', color=registered_color, marker='o', linewidth=2)
casual_peak_hour = casual_by_hr.idxmax()
casual_peak_value = casual_by_hr.max()
plt.annotate('Peak', 
             xy=(casual_peak_hour, casual_peak_value), 
             xytext=(casual_peak_hour, casual_peak_value + 300), 
             arrowprops=dict(facecolor='black', arrowstyle='wedge,tail_width=0.7'))
registered_peak_hour = registered_by_hr.idxmax()
registered_peak_value = registered_by_hr.max()
plt.annotate('Peak', 
             xy=(registered_peak_hour, registered_peak_value), 
             xytext=(registered_peak_hour, registered_peak_value + 300), 
             arrowprops=dict(facecolor='black', arrowstyle='wedge,tail_width=0.7'))
plt.xlabel('Hour', fontweight='bold', fontsize=12)
plt.ylabel('Counts', fontweight='bold', fontsize=12)
plt.title('Casual and Registered Users by Hour (Line Chart)', fontweight='bold', fontsize=16)
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().yaxis.set_tick_params(width=2)
plt.gca().xaxis.set_tick_params(width=2)
plt.show()




```

Interpretasi : Jika dilihat dari grafik di atas, sebenarnya penyewaan casual dan registered cukup selaras. Hal itu dapat 
               dari misalnya pada hour 1, keduanya sama - sama turun, begitu juga pada hour 2. Namun, hal itu hanya terjadi
               jam - jam dini hari. di jam - jam selanjutnya, terutama di jam sore, terdapat perbedaan antara registered dan 
               casual. Peak Hour Casual terjadi pada jam ke-16 dan Peak Hour Registered terjadi pada hour ke-19.
               

### Pertanyaan 2: Bagaimana Kecenderungan Casual dan Registered Renter terhadap Pemilihan Hari Sewa

```python
# Mendefinisikan Data
weekday_labels = list(calendar.day_name)
data_day['weekday_name'] = data_day['weekday'].map(dict(enumerate(weekday_labels)))
data_day_grouped = data_day.groupby('weekday_name')
casual_by_weekday = data_day_grouped['casual'].sum()
registered_by_weekday = data_day_grouped['registered'].sum()

# Membuat Chart
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
plt.rcParams['axes.facecolor'] = "#F0F0F0" 
casual_color = '#2C599D'
registered_color = '#FF5733'
bar_width = 0.4  
bar_positions_casual = range(len(casual_by_weekday))
bar_positions_registered = [pos + bar_width for pos in bar_positions_casual]
casual_bars = plt.bar(bar_positions_casual, casual_by_weekday, color=casual_color, width=bar_width, label='Casual')
registered_bars = plt.bar(bar_positions_registered, registered_by_weekday, color=registered_color, width=bar_width, label='Registered')
for bar, count in zip(casual_bars, casual_by_weekday):
    plt.text(bar.get_x() + bar.get_width() / 2, count, str(count),
             ha='center', va='bottom', fontsize=9, fontweight='bold', color="black")

for bar, count in zip(registered_bars, registered_by_weekday):
    plt.text(bar.get_x() + bar.get_width() / 2, count, str(count),
             ha='center', va='bottom', fontsize=9, fontweight='bold', color="black")
plt.xlabel('Day', fontweight='bold', fontsize=12)
plt.ylabel('Counts', fontweight='bold', fontsize=12)
plt.title('Casual and Registered Users by Weekday', fontweight='bold', fontsize=16)
plt.legend(loc='upper left')
plt.xticks([pos + bar_width / 2 for pos in bar_positions_casual], casual_by_weekday.index)
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().yaxis.set_tick_params(width=2)
plt.gca().xaxis.set_tick_params(width=2)
plt.show()
```

Interpretasi : Dari grafik di atas, dapat diketahui bahwa behavior Casual dan Registered User berbeda jika dilihat dari hari
               sewa. Pada Casual User paling tinggi adalah hari Senin dengan 98.565 user dan terendah pada hari selasa dengan
               52.287 user. Sedangkan, pada Registered User, hari dengan sewa paling tinggi adalah hari rabu dengan 409.065
               dan paling rendah adalah 206.257 user.Dari pernyataan tersebut, tidak dapat diketahui alasan pemilihan hari 
               masing - masing user.

### Pertanyaan 3 : Bagaimana Kecenderungan Casual dan Registered Renter terhadap Musim

```python
season_labels = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
data_day['season_label'] = data_day['season'].map(season_labels)
seasonal_usage = data_day.groupby('season_label')[['registered', 'casual']].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(
    x='season_label',
    y='casual',
    data=seasonal_usage,
    color='#7FC242',  
    label='Casual'
)

sns.barplot(
    x='season_label',
    y='registered',
    data=seasonal_usage,
    color='#83B0C7',  
    label='Registered',
    bottom=seasonal_usage['casual']
)

plt.xlabel('Season', fontweight='bold', fontsize=12)
plt.ylabel('Count', fontweight='bold', fontsize=12)
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Musim', fontweight='bold', fontsize=16)
plt.legend(loc='upper left', title='User Type')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()



```

Interpretasi : Terlihat jelas dari grafik di atas bahwa kecenderungan tipe user (Casual dan Registered) adalah sama berdasarkan
               musim. Hal itu terlihat keduanya sama - sama memiliki tingkat sewa paling tinggi di musim gugur dan tingkat sewa
               paling rendah di musim semi. Begitu juga di musim panas dan dingin, keduanya memiliki kecenderungan proporsi 
               yang sama.
